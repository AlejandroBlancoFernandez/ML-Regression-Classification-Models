```python
import pandas as pd
import pickle
```

```python
# Descarga del fichero
from google.colab import files
files.download('modelo_final.pkl')
```

```python
# Modelo importado
modelo = pickle.load(open('modelo_final.pkl','rb'))
```

```python
comp_df = pd.read_csv("comp_st23ns1.txt.bz2",
compression="bz2",
index_col=0)

comp_df
```

```python
# Utilizamos el fichero entero
predicciones_modelo = modelo.predict(comp_df)
```

```python
# El array NumPy lo convertimos a un dataframe
df_predicciones = pd.DataFrame(predicciones_modelo, columns=['predicciones'])

# Guardamos el dataframe en el fichero
df_predicciones.to_csv('predicciones.csv')
```
